```python
import sys
sys.path.append("../..")

from langchain.llms.ctransformers import CTransformers
```

```python
def load_model():
    llm = CTransformers(
        model="replit-code-v1_5-3b-q4_0.gguf",
        model_type="llama",
        max_new_tokens=1026,
        temperature=0.7,
        repetition_penalty=1.5
    )
    return llm
```

```python
user_input = "Hello"

system_prompt = f"### System:\nYou are an AI assistant named Bob. Help as much as you can."

prompt = f"{system_prompt}### User: {user_input}\n\n### Assistant:\n"
```

```python
llm = load_model()

llm_response = llm(prompt)
llm_response
```
